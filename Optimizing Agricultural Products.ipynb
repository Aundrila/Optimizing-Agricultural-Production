```python
## Lets check the Environmental and Soil Summary of all crops:

print("Average Ratio of Nitrogen in the soil: {0:.2f}".format(data['N'].mean()))
print("Average Ratio of Phosphrous in the soil: {0:.2f}".format(data['P'].mean()))
print("Average Ratio of Pottusium in the soil: {0:.2f}".format(data['K'].mean()))
print("Average Temperature in Celcius: {0:.2f}".format(data['temperature'].mean()))
print("Average Relative Humidity in %: {0:.2f}".format(data['humidity'].mean()))
print("Average PH value of the soil: {0:.2f}".format(data['ph'].mean()))
print("Average Rainfall in mm: {0:.2f}".format(data['rainfall'].mean()))
```

```python
## Lets check the Summary Statistics for each of the crops

@interact
def summary(crops = list(data['label'].value_counts().index)):
    x = data[data['label'] == crops]
    print("---------------------------------------------")
    print("Statistics for Nitrogen")
    print("Minimum required: ", x['N'].min())
    print("Average required: ", x['N'].mean())
    print("Maximum required: ", x['N'].max())
    print("---------------------------------------------")
    print("Statistics for Phosphorous")
    print("Minimum required: ", x['P'].min())
    print("Average required: ", x['P'].mean())
    print("Maximum required: ", x['P'].max())
    print("---------------------------------------------")
    print("Statistics for Potasium")
    print("Minimum required: ", x['K'].min())
    print("Average required: ", x['K'].mean())
    print("Maximum required: ", x['K'].max())
    print("---------------------------------------------")
    print("Statistics for Temparature")
    print("Minimum required: ", x['temperature'].min())
    print("Average required: ", x['temperature'].mean())
    print("Maximum required: ", x['temperature'].max())
    print("---------------------------------------------")
    print("Statistics for Humidity")
    print("Minimum required: ", x['humidity'].min())
    print("Average required: ", x['humidity'].mean())
    print("Maximum required: ", x['humidity'].max())
    print("---------------------------------------------")
    print("Statistics for Ph")
    print("Minimum required: ", x['ph'].min())
    print("Average required: ", x['ph'].mean())
    print("Maximum required: ", x['ph'].max())
    print("---------------------------------------------")
    print("Statistics for Rainfall")
    print("Minimum required: ", x['rainfall'].min())
    print("Average required: ", x['rainfall'].mean())
    print("Maximum required: ", x['rainfall'].max())
    
    
```

```python
## more Intutive

@interact
def compare(conditions = ['N', 'P', 'K', 'temperature', 'ph', 'humidity', 'rainfall']):
    print("Crops which require greater than average ", conditions, '\n')
    print(data[data[conditions] > data[conditions].mean()]['label'].unique())
    print("---------------------")
    print("Crops which require less than average", conditions, '\n')
    print(data[data[conditions] <= data[conditions].mean()]['label'].unique())
```
